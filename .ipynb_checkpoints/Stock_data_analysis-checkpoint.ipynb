```python
from statistics import mean

#to get the average closing price over last n-days
#parameters are number of days and the dataframe
def avg_price(n,df):
    avg_df=mean(df['Close Price'][-n:])
    return avg_df

#to get the profit/loss percentage over last n-days
#parameters are number of days and the dataframe    
def profit_loss_perc(n,df):
    first=df['Close Price'].iloc[-n]
    second=df['Close Price'].iloc[-1]
    num=first-second
    denom=abs(first)
    profit_loss=(num/denom)*100
    return profit_loss
            
# getting the average and profit/loss percentage over the last
# 1 week=7 days
# 2 weeks=14 days
# 1 month=30 days
# 3 months=90 days
# 6 months=180 days
# 1 year=365 days
no_days=[7,14,30,90,180,365]
for i in no_days:
    print('Average of last {} days:'.format(i))
    print(avg_price(i,df))
    print('Profit/Loss percent of last {} days:'.format(i))
    print(profit_loss_perc(i,df))
    print('-'*20)
```

```python
#Adding new row with daily percentage changes values
df['Day_Perc_Change']=df['Close Price'].pct_change()
```

```python
df.head()
```

```python
#First value is Nan..Changing it to 0
df['Day_Perc_Change']=df['Day_Perc_Change'].fillna(0)
```

```python
df.head()
```

```python
def trends(perc): 
    if(perc>-0.5 and perc<0.5):
        return 'Slight or No change'
        
    elif(perc>0.5 and perc<1):
        return 'Slight positive'
    
    elif(perc>-1 and perc<-0.5):
        return 'Slight negative'
        
    elif(perc>1 and perc<3):
        return 'Positive'
        
    elif(perc>-3 and perc<-1):
        return 'Negative'
        
    elif(perc>3 and perc<7):
        return 'Among top gainers'
        
    elif(perc>-7 and perc<-3):
        return 'Among top losers'
        
    elif(perc>7):
        return 'Bull run'
        
    elif(perc<-7):
        return 'Bear drop'
    
df['Trends']=df['Day_Perc_Change'].map(trends)
```

```python
df.head()
```

```python
grp_t=df.groupby(by=['Trends'])

tmean=grp_t['Total Traded Quantity'].mean()
tmedian=grp_t['Total Traded Quantity'].median()

print(tmean)
print(tmedian)
```
