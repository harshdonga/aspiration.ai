```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
```

```python
df = pd.read_csv('./data/GOLD.csv')
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
```

```python
#lets see pred values over range given
plt.figure(figsize=(25,15))
plt.subplot(212)
plt.plot(df['Pred'][:411], color='lightseagreen', label= 'Pred')
```

```python
x = df[['Price', 'Open', 'High', 'Low']]
y = df['Pred']
model = LinearRegression()
model.fit(x[:411],y[:411])
print('Intercept: {}'.format(model.intercept_))
print('Coefficients: \nClose: {}\nOpen: {}\nHigh: {}\nLow: {}'.format(int(model.coef_[0]),int(model.coef_[1]),int(model.coef_[2]),int(model.coef_[3])))
```

```python
df['Pred'][411:] = model.predict(x[411:])
```

```python
plt.figure(figsize=(25,15))
plt.subplot(212)
plt.plot(df['Pred'][:411], color='lime')
plt.plot(df['Pred'][411:], color='r')
```

```python
model_2 = LinearRegression()
model_2.fit(x[:411], df['new'][:411])
print('Intercept: {}'.format(model_2.intercept_))
print('Coefficients: \nClose: {}\nOpen: {}\nHigh: {}\nLow: {}'.format(model_2.coef_[0],model_2.coef_[1],model_2.coef_[2],model_2.coef_[3]))
```

```python
df['new_'] = df['new']
df['new_'][411:] = model_2.predict(x[411:])
```

```python
plt.figure(figsize=(25,15))
plt.subplot(211)
plt.plot(df['new'], color='lightseagreen')
plt.subplot(212)
plt.plot(df['new'], color='lightseagreen')
plt.plot(df['new_'][411:], color='aquamarine')
```

```python
df_cipla = pd.read_csv('./data/Large_Cap/CIPLA.csv')
df_nifty = pd.read_csv('./data/Nifty50/Nifty50.csv')
```

```python
df_cipla.info()
```

```python
df_betas = pd.concat([df_cipla['Date'][:494],df_cipla['Close Price'][:494], df_nifty['Close'][:494]],axis=1)
```

```python
df_betas.columns=['Date','CIPLA','NIFTY']
```

```python
df_betas_pct = df_betas[['CIPLA', 'NIFTY']].pct_change(1)[1:]
```

```python
#for last three months:

import statsmodels.api as sm
x = df_betas_pct['NIFTY'][403:]
y = df_betas_pct['CIPLA'][403:]
X1 = sm.add_constant(x)
model = sm.OLS(y, X1)
results = model.fit()
```

```python
print(results.summary2())
```

```python
results.params[1]
```

```python
### DAILY BETA PRED OF LAST 90 DAYS ###
daily_beta_90 = results.predict(X1)
```

```python
df_betas['Month'] = pd.DatetimeIndex(df_betas['Date']).month
```

```python
month_grouped = df_betas.groupby('Month')
```

```python
month_grouped.first()
```

```python
### from may 2017 to apr 2018 ###

for i , j in month_grouped:
    j_pct = j[['CIPLA', 'NIFTY']].pct_change(1)[1:]
    x = j_pct['NIFTY']
    y = j_pct['CIPLA']
    X1 = sm.add_constant(x)
    model = sm.OLS(y, X1)
    results = model.fit()
    print("For Month No {}, Beta value : {}".format(i,results.params[1]))
```
