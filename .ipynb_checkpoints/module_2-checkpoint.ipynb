```python
import pandas as pd
import numpy as np
%matplotlib inline 
import matplotlib.pyplot as  plt
import seaborn as sns
import matplotlib
matplotlib.use('nbagg')
sns.set_style("darkgrid")
```

```python
df = pd.read_csv('./results/week2.csv')
df.head()
```

```python
df['Date'] = pd.to_datetime(df.Date) 
df.info()
```

```python
x1=df['Date'].tolist()
y1=df['Day_Perc_Change'].tolist()
y2=df['Total Traded Quantity'].tolist()


plt.figure(1)

plt.subplot(211)
plt.ylabel("Date")
plt.xlabel('Day Percentage Change')
plt.stem(x1,y1,'b')


plt.subplot(212)
plt.xlabel("Date")
plt.ylabel('Volume')
plt.plot(x1,y2,'r')
plt.figure(figsize=(25,15))
plt.show()
```

```python
print('Since all our data points lie in same group ie No Change\nPerforming groupby query would be trivial\n\n')

trends_group = df.groupby(by='Trend')
trends_group.first()
```

```python
plt.figure(2)
labels=['No change']
plt.pie(trends_group['Date'].agg('count'),labels=labels, autopct='%0.1f%%')
```

```python
plt.figure(4)
plt.subplot(211)
avg_trends=trends_group.agg('mean')
avg_trends['TTQ'].plot(kind='bar',title='Mean of volume in trends',color='r')

plt.subplot(212)
median_trends=trends_group.agg('median')
median_trends['TTQ'].plot(kind='bar',title='Median of volume in trends',color='b')
```

```python
df_CP=pd.concat([df['Close Price'],df_SP['Close Price'],df_ITC['Close Price'],df_HDFC['Close Price'],df_TCS['Close Price'], df_MM['Close Price']], axis=1)
```

```python
df_CP = df_CP[:503]
df_CP.columns = ['AMZN', 'SUN', 'ITC', 'HDFC', 'TCS', 'MM']
```

```python
df_CP_pct =  df_CP.pct_change()[1:]
df_CP_pct.head(4)
```

```python
df_corr=df_CP_pct.corr(method='pearson')
plt.figure(4)
sns.pairplot(df_corr, height=3);
plt.show()
```

```python
df_AMZN_roll_vol = df_CP_pct['AMZN'].rolling(7).std()
```

```python
plt.figure(5)
plt.plot(df_AMZN_roll_vol)
plt.show()
```

```python
df_nifty=pd.read_csv('./data/Nifty50/Nifty50.csv')
df_nifty=df_nifty['Close'].pct_change()
df_nifty_roll_vol=df_nifty.rolling(7).std()
```

```python
plt.figure(6)
plt.plot(df_AMZN_roll_vol,label='AMZN',color='b')
plt.plot(df_nifty_roll_vol,label='Nifty',color='r')
plt.legend()
plt.show()
```

```python
def call_trade(n1,n2,df,Day):
    df_roll=df.rolling(n1).mean()
    df_roll_2=df.rolling(n2).mean()
    plt.figure(6)
    plt.plot(df_roll,label=str(n1) + 'SMA',color='r')
    plt.plot(df_roll_2,label=str(n2) + 'SMA',color='b')
    plt.legend()
    plt.show()
    if(df_roll.iloc[Day] >df_roll_2.iloc[Day] ):
        return 'BUY'
    else:
        return 'SELL'
call_trade(21,34,df_CP['AMZN'],15)
```

```python
df_roll_avg=df_CP['AMZN'].rolling(14).mean()
df_std=df_CP['AMZN'].std()
df_upper=df_roll_avg+(2*df_std)
df_lower=df_roll_avg-(2*df_std)

plt.figure(7)
plt.plot(df_roll_avg,label='Rolling mean',color='r')
plt.plot(df_upper,label='Bollinger upper band',color='b')
plt.plot(df_lower,label='Bollinger lower band',color='g')
plt.plot(df_CP['AMZN'],label='Average Daily Stock',color='black')
plt.legend()
plt.show()
```

```python
df['Rolling_14']=df_roll_avg
df['ubollinger']=df_upper
df['dbollinger']=df_lower
```

```python
df.drop('Unnamed: 0', axis=1)
```

```python
df.to_csv('week3.csv', index = False)
```
