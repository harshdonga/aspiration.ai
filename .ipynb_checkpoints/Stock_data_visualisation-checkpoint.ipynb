```python
df_nifty=pd.read_csv('Nifty50//Nifty50.csv')
df_nifty=df_nifty['Close'].pct_change()
df_nifty_roll=df_nifty.rolling(7)
df_nifty_vol=df_nifty_roll.std()
```

```python
def call_buy_sell(n1,n2,df,Day):
    df_roll=df.rolling(n1).mean()
    df_roll_2=df.rolling(n2).mean()
    plt.figure(6)
    plt.plot(df_roll,label=str(n1) + 'SMA',color='r')
    plt.plot(df_roll_2,label=str(n2) + 'SMA',color='b')
    plt.legend()
    plt.show()
    
    first=df_roll.iloc[Day]
    second=df_roll_2.iloc[Day]
    if(first>second):
        return 'BUY'
    else:
        return 'SELL'
    
    
call_buy_sell(21,34,df_close['RCOM'],15)
```

```python
df_roll_avg=df_close['RCOM'].rolling(14).mean()
df_std=df_close['RCOM'].std()
df_upper=df_roll_avg+(2*df_std)
df_lower=df_roll_avg-(2*df_std)

plt.figure(7)
plt.plot(df_roll_avg,label='Rolling mean',color='r')
plt.plot(df_upper,label='Bollinger upper band',color='b')
plt.plot(df_lower,label='Bollinger lower band',color='g')
plt.plot(df_close['RCOM'],label='Average Daily Stock',color='black')
plt.legend()
plt.show()
```

```python
df['Rolling Mean 14 Days']=df_roll_avg
df['BOLU']=df_upper
df['BOLD']=df_lower
```

```python
df.to_csv('week3.csv',index=False)
```
