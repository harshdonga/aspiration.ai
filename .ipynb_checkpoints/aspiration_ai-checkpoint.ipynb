```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as ss
```

```python
df = pd.read_csv('./data/AMZN.csv')
```

```python
df.info()
```

```python
pd.isnull(df).any()
```

```python
df.dtypes
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
for i in df.columns[1:-1]:
    df[i] = df[i].str.replace(',', '').astype(float)
```

```python
df['TTQ'] = df['Total Traded Quantity'].str.replace('M','').astype(float).apply(lambda x : x*10**6).astype(int)
```

```python
sns.set_style("darkgrid")
df.plot(x='Date', y=['Close Price', 'Open Price', 'High Price', 'Low Price'], grid=True, figsize=(25,15))
```

```python
max_90 = max(df['Close Price'][:90])
min_90 = min(df['Close Price'][:90])
mean_90 =  ss.mean(df['Close Price'][:90])
print("Task 1.2\n\n")
print("Maximum Close Price : {} \nMinimum Close Price : {} \nMean Close Price : {}".format(max_90,min_90,mean_90))
```

```python
df['Month'] = pd.DatetimeIndex(df['Date']).month
df['Year'] = pd.DatetimeIndex(df['Date']).year
```

```python
df['PxV'] = df['Close Price'] * df['TTQ']
```

```python
month_grouped = df.groupby('Month')
year_grouped = df.groupby('Year')
```

```python
vwap = []
for i , j in month_grouped:
    vwap.append(sum(j['PxV'])/sum(j['TTQ']))
    
print("Task 1.4\n\n")
print(vwap)
```

```python
def calculate_avg(n):
    return(df['Close Price'][:n].mean())


def calculate_profit_loss(n):
    return(((df['Close Price'][n] - df['Close Price'][0])/df['Close Price'][n])*100)
    
print("Task 1.5\n\n")
print("Average Price over Last:\n\n1 week : {}\n2 weeks: {}\n1 month: {}\n3 months: {}\n6 months: {}\n1 year: {}".format(calculate_avg(7),calculate_avg(14), calculate_avg(30),calculate_avg(91), calculate_avg(183),calculate_avg(365)))
print("\n\nProfit/Loss % over Last:\n\n1 week : {}\n2 weeks: {}\n1 month: {}\n3 months: {}\n6 months: {}\n1 year: {}".format(calculate_profit_loss(7),calculate_profit_loss(14), calculate_profit_loss(30),calculate_profit_loss(91), calculate_profit_loss(183),calculate_profit_loss(365)))
```

```python
df['Day_Perc_Change'] = df['Close Price'].pct_change()
```

```python
df['Day_Perc_Change'][0] = 0
print("Task 1.6\n\n")
df.head()
```

```python
def trend(x):
    if x['Day_Perc_Change'] > -0.5 and x['Day_Perc_Change'] <= 0.5:
        return 'No Change'
    elif x['Day_Perc_Change'] > -1 and x['Day_Perc_Change'] <= -0.5:
        return 'Slight Negative'
    elif x['Day_Perc_Change'] > 0.5 and x['Day_Perc_Change'] <= 1:
        return 'Slight Positive'
    elif x['Day_Perc_Change'] > 1 and x['Day_Perc_Change'] <= 3:
        return 'Positive'
    elif x['Day_Perc_Change'] > -3 and x['Day_Perc_Change'] <= -1:
        return 'Negative'
    elif x['Day_Perc_Change'] > 3 and x['Day_Perc_Change'] <= 7:
        return 'Among Top Gainers'
    elif x['Day_Perc_Change'] > -7 and x['Day_Perc_Change'] <= -3:
        return 'Among Top Losers'
    elif x['Day_Perc_Change'] > 7:
        return 'Bull Run'
    else:
        return 'Bear Drop'

df['Trend'] = df.apply(lambda x: trend(x), axis= 1)
```

```python
trend_grouped = df.groupby('Trend')
```

```python
print('Since only one trend!')
print('Average: {}\n Median: {}'.format(df['TTQ'].mean(),df['TTQ'].median() ))
```
